```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import warnings
warnings.simplefilter("ignore")
import time
import os
import random
import json

headers={
    
    
}

def get_source_html(url):    
    try:
        options = webdriver.ChromeOptions()
        options.add_experimental_option("detach", True)
        driver = webdriver.Chrome(chrome_options=options,executable_path=r'C:\Users\svnduw\.wdm\drivers\chromedriver\win32\95.0.4638.54\\chromedriver.exe')
        driver.maximize_window()
        driver.get('https://sbis.ru/contragents?p=contragents')
        time.sleep(2)
        driver.find_element_by_class_name('controls-EditorList__footer').click()
        time.sleep(2)
        driver.find_elements_by_id('RegionCityPanel_region')[84].click() #region name
        time.sleep(2)
        driver.find_element_by_class_name('sbis_ru-CookieAgreement__close').click()
        time.sleep(1)
        driver.find_element_by_xpath('//*[@id="container"]/div[2]/div/div/div/div[2]/div[1]/div/div/div/div[2]/div/div[1]/div[2]/div/div/div/div[1]/div[4]/div/div/div[2]/div[1]/div/div[6]/div[2]/div/div/div').click()
        time.sleep(2)
        driver.find_element_by_xpath('//*[@id="popup"]/div[2]/div/div[2]/div/div/div/div[1]/div/div/div[2]/div[1]/div/div[3]/div[20]/div[2]/div/div[1]').click()
        time.sleep(2)
        driver.find_element_by_xpath('//*[@id="popup"]/div[2]/div/div[2]/div/div/div/div[1]/div/div/div[2]/div[1]/div/div[3]/div[25]/div[2]/div/div[2]').click()
        count=0
        while True:
            numbers_down=driver.find_elements_by_xpath('//*[@id="container"]/div[2]/div/div/div/div[2]/div[1]/div/div/div/div[2]/div/div[2]/div[3]/div/div/div/span[3]')
            if len(numbers_down)==1 or 0:
                count+=1
                with open(r'C:\Users\svnduw\Desktop\Bekbol\Learning\freelance\html\source'+'_'+str(count)+'.html','w', encoding="utf-8") as file:
                    file.write(driver.page_source)
                numbers_down[0].click()
                time.sleep(4)
            else:        
                break
    except Exception as _ex:
        print(_ex)
    finally:
        driver.close()
        driver.quit()
    return "[INFO] collected html files"




def get_item_urls(file_path):
    files = os.listdir(file_path)
    all_iin=[]
    for file in files:
        with open(file_path+'/'+file, encoding="utf-8") as f:
            src=f.read()
            soup=BeautifulSoup(src,"lxml")
            iins=soup.find_all('div',class_='controls-Grid__row-cell__content js-controls-ListView__editingTarget controls-Grid__row-cell__content_baseline_default controls-Grid__row-cell_cursor-pointer controls-Grid__cell_spacingLeft controls-Grid__cell_spacingRight controls-Grid__row-cell_rowSpacingTop_l controls-Grid__row-cell_rowSpacingBottom_l controls-Grid__row-cell_withoutRowSeparator_size-null controls-Grid__item_background-hover_default sbis-Contragents-ContragentsList__Inn')
            for iin in iins:              
                all_iin.append(iin.text)
    all_iin=list(set(all_iin))
    with open(r'C:\Users\svnduw\Desktop\Bekbol\Learning\freelance\iin\iins.txt','w') as file:
        for iin in all_iin:
            file.write('https://sbis.ru/contragents/'+f"{iin}\n")
        print( "[INFO] iins collected successfully!")
        
        
        
        
        
def get_item_data(file_path):
    
    
    
    
    with open(file_path) as file:
        url_list=file.readlines()
        clear_url=[]
        for url in url_list:
            clear_url.append(url.strip())
            
            
        result_list=[]
        count_sleep=0
        url_count=len(clear_url)
        
        
        
        for url in clear_url:
            names=[]
            directors=[]
            positions=[]
            employes=[]
            addresses=[]
            phones=[]
            revenues=[]
            profits=[]
            costs=[]
            response=requests.get(url=url, headers=headers)
            soup=BeautifulSoup(response.text,'lxml')
            
            try:
                name=soup.find("div",{'itemprop':'name'}).text.strip()
                names.append(name) 
            except Exception as ex:
                name=None
                names.append(name)
                
            try:
                director=soup.find("span",{'itemprop':'employee'}).text
                directors.append(director) 
            except Exception as ex:
                director=None
                directors.append(director)
                
            try:
                position=soup.find("div",class_='cCard__Director-Position').text
                positions.append(position) 
            except Exception as ex:
                position=None
                positions.append(position)
            
            try:
                employe=soup.find("span",{'itemprop':'numberOfEmployees'}).text
                employes.append(employe) 
            except Exception as ex:
                employe=None
                employes.append(employe)
                
            try:
                address=soup.find("div",{'itemprop':'address'}).text
                addresses.append(address) 
            except Exception as ex:
                address=None
                addresses.append(address)
                
            try:
                phone=soup.find("div",{'itemprop':'telephone'}).text
                phones.append(phone) 
            except Exception as ex:
                phone=None
                phones.append(phone)
                
                
            all_prices=soup.find_all("div",class_='cCard__BlockTitle cCard__Padding10')
            try:
                revenue=all_prices[0].find('span',class_='cCard__BlockMaskSum').text
                revenues.append(revenue)
                profit=all_prices[1].find('span',class_='cCard__BlockMaskSum').text
                profits.append(profit)
                cost=all_prices[2].find('span',class_='cCard__BlockMaskSum').text
                costs.append(cost)
            except Exception as ex:
                revenue=None
                revenues.append(revenue)
                profit=None
                profits.append(revenue)
                cost=None
                costs.append(revenue) 
            
            result_list.append( {
                
                
                'Название компании': names,
                'Адресс': addresses,
                'CEO':directors,
                'Позиция':positions,
                'Телефон': phones,
                'Количество сотрудников':employes,
                'Выручка':revenues,
                'Прибыль':profits,
                'Стоимость':costs,               
                'Cсылка на сайт':url            
                
            
            })
            
            time.sleep(random.randrange(2,5))
            
            if count_sleep%10==0:
                time.sleep(random.randrange(5,9))
            count_sleep+=1
            print(f"[+] Processed: {count_sleep}/{url_count}")
            
        with open(r'C:\Users\svnduw\Desktop\Bekbol\Learning\freelance\iin\result.json','w',encoding='utf-8') as file:
            json.dump(result_list,file,indent=4,ensure_ascii=False)
        return "[INFO] Collected successfully!"
            
        
            
                
            
        
                
            
def main():
    #get_source_html('https://sbis.ru/contragents?p=contragents')
    #get_item_urls(r'C:\Users\svnduw\Desktop\Bekbol\Learning\freelance\html')
    get_item_data(r'C:\Users\svnduw\Desktop\Bekbol\Learning\freelance\iin\iins.txt')
        
if __name__=="__main__":
     main()
        
        
        
```
